# BULLET POINT 1: Remove rows with NAs

```python
ai.isna().sum()
```

```python
ai = ai.dropna()
ai.isna().sum()
```

# BULLET POINT 2 & 3: Remove the row with the single outlier in num_accts. Filter with a logical condition. Remove the rows with negative income values. Filter with a logical condition.

```python
ai.describe()
```

```python
ai = ai[(ai['income'] > 0) &
              (ai['num_accts'] < 10)]
```

```python
ai.describe()
```

# BULLET POINT 4: Make the 4 variables you have identified in the previous question into categorical variables.

```python
ai.info()
```

```python
ai['job'] = ai['job'].astype('category')
ai['product'] = ai['product'].astype('category')
ai['chk_acct'] = ai['chk_acct'].astype('category')
ai['sav_acct'] = ai['sav_acct'].astype('category')
```

```python
ai.info()
```

# Calculate and report the mean of "answered" after cleaning the dataset.  Because "answered" is coded 0/1 this is as simple as calculating the proportion of 1s in that series with .mean().

```python
mean_answered = ai['answered'].mean()*100
print(mean_answered)
```

# PLOT 1


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ai['answered'].value_counts()
```

```python
ai['answered'] = ai['answered'].replace({1: 'Yes', 0: 'No'})
ai['answered'] = ai['answered'].astype('category')
```

```python
ai['answered'].value_counts()
```

```python
sns.boxplot(x='answered', y='income', data=ai).set(title = 'Income by Answered', xlabel = 'Answered', ylabel = 'Income')
```

# PLOT 2


```python
ai['mobile'] = ai['mobile'].replace({0: 'No', 1: 'Yes'})
ai['mobile'] = ai['mobile'].astype('category')
```

```python
ai.groupby(['mobile', 'answered'])['answered'].count().unstack()
```

```python
ai.groupby(['mobile', 'answered'])['answered'].count().unstack().plot(
    kind='bar', title = 'Count of Answered by Mobile', rot = 0, xlabel = 'Mobile', ylabel = 'Counts')
```

```python
(ai.groupby(['mobile', 'answered'])['answered'].count().unstack().apply(lambda x: x/x.sum(), axis = 1)).plot(
    kind='bar', title = 'Proportion of Answered by Mobile Status', rot = 0, xlabel = 'Mobile', ylabel = 'Proportion')
```

# What is the accuracy of the income model?

```python
money_tree = DecisionTreeClassifier(criterion = "entropy", max_depth = 2)
```

```python
money_tree = money_tree.fit(X = ai[['income']],
                            y = ai['answered'])
```

```python
pred = money_tree.predict(X = ai[['income']])
```

```python
round(sum((pred == ai['answered']) / len(ai)),3)
```
